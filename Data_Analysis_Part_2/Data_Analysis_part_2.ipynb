Importing libraries (Numpy, pandas, matplotlib, plotly and seaborn) and then read csv file.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as po
```

Checking statistical facts by checking all columns. Then calculate the maximum value of the following
attributes ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (%

```python
column = ['Calories', 'Total Fat', 'Carbohydrates','Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']
```

```python
for col in column:
  a = (df2[df2[col] == (df2[col].max())]['Item'].values[0])
  print(f"{col} : {a} - {(df2[col].max()).astype(float)}")
```

```python
plt.figure(figsize = (30, 10))
sns.set(style = "whitegrid")
plt.subplot(5, 2,1)

plt.title("Calaries Table")
sns.stripplot(data =df2, x='Category', y='Calories', palette = "bright")
plt.xticks(rotation = 45)

plt.title("Total Fat Table")
plt.subplot(5, 2 ,2)
sns.stripplot(data =df2, x='Category', y='Total Fat', palette = "bright")
plt.xticks(rotation = 45)

plt.title("Carbohydrates Table")
plt.subplot(5, 2  ,3)
sns.stripplot(data =df2, x='Category', y='Carbohydrates', palette = "bright")
plt.xticks(rotation = 45)

plt.title("Dietary Fiber Table")
plt.subplot(5, 2 ,4)
sns.stripplot(data =df2, x='Category', y='Dietary Fiber', palette = "bright")
plt.xticks(rotation = 45)

plt.title("Sugars Table")
plt.subplot(5, 2  ,5)
sns.stripplot(data =df2, x='Category', y='Sugars', palette = "bright")
plt.xticks(rotation = 45)

plt.title("Protein Table")
plt.subplot(5, 2 , 6)
sns.stripplot(data =df2, x='Category', y='Protein', palette = "bright")
plt.xticks(rotation = 45)

plt.title("Vitamin A (% Daily Value) Table")
plt.subplot(5, 2  ,7)
sns.stripplot(data =df2, x='Category', y='Vitamin A (% Daily Value)' , palette = "bright")
plt.xticks(rotation = 45)

plt.title("Vitamin C (% Daily Value) Table")
plt.subplot(5, 2 , 8)
sns.stripplot(data =df2, x='Category', y='Vitamin C (% Daily Value)', palette = "bright")
plt.xticks(rotation = 45)

plt.title("CalaCalcium (% Daily Value) Table")
plt.subplot(5, 2 ,9)
sns.stripplot(data =df2, x='Category', y='Calcium (% Daily Value)', palette = "bright")
plt.xticks(rotation = 45)

plt.title("CalarIron (% Daily Value)ies Table")
plt.subplot(5, 2 , 10)
sns.stripplot(data =df2, x='Category', y='Iron (% Daily Value)', palette = "bright")
plt.xticks(rotation = 45)

plt.show()
```
