```python
x= np.array([10,20,  30])
y= np.array([20, 40 ,10])
```

```python
plt.plot(x, y, color='blue')
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Sample Graph")
plt.xlim(10, max(x))
plt.show()
```

```python
x2= np.array([10,20,  30])
y2= np.array([40, 10 ,30])
```

```python
plt.plot(x, y, color='blue')
plt.xlim(10, max(x))


plt.plot(x2, y2, color='red')
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Sample Graph")
plt.xlim(10, max(x))
plt.ylim(10,40)
plt.show()
```
